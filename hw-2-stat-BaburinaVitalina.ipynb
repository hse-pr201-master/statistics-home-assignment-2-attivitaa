```python
#Номер 1а

#Вероятность встретить нового таксиста в первый день - Р1= 1
#Вероятность встретить нового таксиста во второй день - P2= 1 * (n-1)\n
#Вероятность встретить нового таксиста в к день - Pk= 1* (n-1)\n * (n-2)\n * ...* (n-9)\n
#Вероятность встретить таксиста, который уже приезжал на х заказе: Px= 1* (n-1)\n * (n-2)\n * ...* 9\n
import numpy as np
import matplotlib.pyplot as plt
n=np.arange(1, 800)
def L1(n):
    L1=1
    for i in range (9):
        L1=L1*(n-i)/n
    L1=L1*9/n
    return (L1)
plt.plot(n, L1(n))
plt.axvline(np.argmax(L1(n)), color='g',label='Метод максимального правдоподобия')
plt.legend()
print('Оценка методом максимального правдоподобия:')
print (np.argmax(L1(n))+1)
```
